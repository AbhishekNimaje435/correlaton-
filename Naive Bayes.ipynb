# Naive Bayes Model

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\HP\Downloads\placement-dataset.csv")
print(data.head(3))
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x="cgpa",y="score",hue="placement",data=data)
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sns.kdeplot(data=data["cgpa"])
plt.show()
```

```python
sns.kdeplot(data=data["score"])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

# Gaussian Naive Bayes

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100 , gnb.score(x_train,y_train)*100
```

```python
from mlxtend.plotting import plot_decision_regions
pdg=plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

# Multinomial Naive Bayes

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100 , mnb.score(x_train,y_train)*100 
```

```python
from mlxtend.plotting import plot_decision_regions
pdg=plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100 , bnb.score(x_train,y_train)*100
```

```python
from mlxtend.plotting import plot_decision_regions
pdg=plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
