```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\HP\Downloads\tips (1).csv")
print(data.head(3))
```

```python
data.info()
```

```python
corr_data=data.select_dtypes(["float64","int64"]).corr()
corr_data
```

```python
cov_data=data.select_dtypes(["float64","int64"]).cov()
cov_data
```

```python
sns.heatmap(corr_data,annot=True)
plt.show()
```

```python
sns.heatmap(cov_data,annot=True)
plt.show()
```
