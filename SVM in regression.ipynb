```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
```

```python
data=pd.read_csv(r"C:\Users\HP\Downloads\placement.csv")
print(data.head(3))
```

```python
x=data.iloc[:,:-1]
y=data["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
sv=SVR(kernel='linear')
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```
