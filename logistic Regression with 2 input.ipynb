# Logistic Regression With Dual Input

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\HP\Downloads\Social_Network_Ads.csv")
#print(data)
data.drop(columns=["User ID"],inplace=True)
data.drop(columns=["EstimatedSalary"],inplace=True)
data.drop(columns=["Gender"],inplace=True)
print(data)
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x="Age",y="Purchased",data=data)
plt.show()
```

```python
x=data[["Age"]]
y=data["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
print(lr.predict([[46]]))
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x="Age",y="Purchased",data=data)
sns.lineplot(data=data,x="Age",y=lr.predict(x),color="black")
plt.show()
```
